```python
# Print all even numbers from 1 to 50

# Loop through numbers 1 to 50
for number in range(1, 51):
    # Check if the number is even (divisible by 2 with no remainder)
    if number % 2 == 0:
        print(number)
```

```python
import pandas as pd
import os

# Find your diabetes CSV file in Downloads
downloads_folder = os.path.expanduser('~/Downloads')

# Look for any CSV with 'diabetes' in the name
for file in os.listdir(downloads_folder):
    if 'diabetes' in file.lower() and file.endswith('.csv'):
        file_path = os.path.join(downloads_folder, file)
        print(f"Found: {file}")
        break

# Load it
df = pd.read_csv(file_path)
print("\nDataset loaded successfully!")
print(df.head())
```

```python
# Check the shape (rows, columns)
print(f"Dataset has {df.shape[0]} rows and {df.shape[1]} columns")

# See column names
print("\nColumns:", df.columns.tolist())

# Check for missing values
print("\nMissing values:\n", df.isnull().sum())

# Basic statistics
print("\nStatistics:\n", df.describe())

# Check the outcome distribution (how many have diabetes vs don't)
print("\nDiabetes cases:\n", df['Outcome'].value_counts())
```
